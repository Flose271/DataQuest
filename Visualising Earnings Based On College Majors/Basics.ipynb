```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 

recent_grads = pd.read_csv("recent-grads.csv")
#firstrow = recent_grads.iloc[0,:]
#print(firstrow)
#recent_grads.head(5)
#recent_grads.describe()
raw_data_count = recent_grads.iloc[:,0].size
print(raw_data_count)
recent_grads = recent_grads.dropna()
raw_data_count = recent_grads.iloc[:,0].size
print(raw_data_count)
```

Only one row was dropped.

```python
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter')
```

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Median vs Sample size')
```

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Unemployment rate vs. Sample size')
```

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Full time employees vs. Median')
```

```python
ax = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax.set_title('Men vs Median')
```

```python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax.set_title('Women vs Median')
```

```python
ax = recent_grads.plot(x='ShareWomen', y='Median', kind='scatter')
ax.set_title('Share of women vs Median')
```

As the share of women increases, the median earnings of graduates from that major slightly decrease.
