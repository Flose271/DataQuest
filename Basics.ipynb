```python
%%sql

SELECT g.name as "Genre", COUNT(t.track_id) as "Sold"
FROM invoice i
INNER JOIN invoice_line il ON il.invoice_id = i.invoice_id
INNER JOIN track t ON t.track_id = il.track_id
INNER JOIN genre g ON g.genre_id = t.genre_id
WHERE i.billing_country = 'USA'
GROUP BY g.genre_id
```

```python
%%sql

WITH customer_totals AS 
(
SELECT c.customer_id, c.support_rep_id, SUM(i.total) AS total
FROM customer c
INNER JOIN invoice i ON i.customer_id = c.customer_id
GROUP BY c.customer_id
)

SELECT e.last_name AS "Surname", 
       e.first_name AS "First name",
       SUM(ct.total) AS "Total Sold"
FROM customer_totals ct
INNER JOIN employee e ON e.employee_id = ct.support_rep_id
GROUP BY e.employee_id
```

```python
%%sql

WITH countries AS
(
SELECT i.billing_country AS "Country", 
       COUNT(DISTINCT i.customer_id) AS "Customers",
       SUM(i.total) AS "Total Sales",
       SUM(i.total)/
           CAST(COUNT(DISTINCT i.customer_id) AS float)
           AS "Sales per customer",
       AVG(i.total) AS "Average Sale",
       CASE WHEN COUNT(DISTINCT i.customer_id) = 1 THEN 1
            ELSE 0
            END AS sort
FROM invoice i
INNER JOIN invoice_line il ON il.invoice_id = i.invoice_id
GROUP BY i.billing_country
)

SELECT "Country", "Customers", "Total Sales", "Sales per customer",
       "Average Sale"
FROM countries
ORDER BY sort ASC, "Total Sales" DESC
```
